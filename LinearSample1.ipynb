```python
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
Linear = LinearRegression()
```

```python
X = [[1],[2],[3],[4],[5]]
Y = [[12],[25],[34],[48],[61]]
```

```python
#In this case, it draws the graph with the original records
plt.title("My graph")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.scatter(X,Y)
plt.plot(X,Y)
```

```python
Model=Linear.fit(X,Y)
```

```python
#After the learning process, the prediction model is made
plt.title("My graph")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(X,Model.coef_*X + Model.intercept_)
plt.scatter(X,Model.coef_*X+Model.intercept_)
```

```python
Model.predict([[12]])
#number is floating point coz the x values in the first place are not perfectly divided
```
