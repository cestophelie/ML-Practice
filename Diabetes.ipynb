```python
from sklearn.linear_model import LinearRegression
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
print(diabetes.feature_names)
```

```python
print(diabetes.data.shape)
#(samples total, dimension_NumOfFeatures)
```

```python
x_axis = diabetes.data #data refering to the independent variables
y_axis = diabetes.target
df = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
print(df)
```

```python
plt.title("Diabetes Graph")
i = 2 #age feature, the second column is sex value so not appropriate for the linear regression test
plt.xlabel(diabetes.feature_names[i])
plt.ylabel('Target values')
plt.scatter(df[diabetes.feature_names[i]],y_axis)
#plt.legend('Range of age')
```

```python
Linear = LinearRegression()
Model = Linear.fit(x_axis,y_axis)

plt.title("Diabetes Graph")
i = 2
plt.xlabel(diabetes.feature_names[i])
plt.ylabel('Target values')
plt.scatter(df[diabetes.feature_names[i]],y_axis)
plt.plot(df[diabetes.feature_names[i]],Model.coef_[i]*df[diabetes.feature_names[i]],'r-')
plt.show()
```
