```python
from sklearn.linear_model import LinearRegression
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
cali = fetch_california_housing()
```

```python
X_axis = cali.data #data refering to the independent variables
dataFrame = pd.DataFrame(cali.data,columns=cali.feature_names)
Y_axis = cali.target #target refers to the value I am trying to get. Often a predicted value
print(dataFrame)
```

```python
#there has to be corresponding one feature-horizontal axis regarding the vertical axis
plt.title("My graph")
plt.xlabel(cali.feature_names[4])
plt.ylabel("Values")
plt.scatter(dataFrame[cali.feature_names[4]],Y_axis)
#scatter literally shows the exact corresponding values whereas plot is connecting all the values
plt.legend()
plt.show()
#later add the functions of the two methods above, legend() and show()
```

```python
#Again, keep in mind one feature corresponds to one target axis
idx = 2
Linear = LinearRegression()
Model = Linear.fit(X_axis,Y_axis)

plt.title("My graph")
plt.xlabel(cali.feature_names[idx])
plt.ylabel("Values")
plt.scatter(dataFrame[cali.feature_names[idx]],Y_axis)
plt.plot(dataFrame[cali.feature_names[idx]],Model.coef_[idx]*dataFrame[cali.feature_names[idx]],'r-')
plt.legend()
plt.show()
```
